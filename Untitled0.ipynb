```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

rng = np.random.RandomState(42)
X = 5 * rng.rand(80, 1)
y = np.sin(X).ravel() + 0.35 * rng.randn(80)

model = LinearRegression()
model.fit(X, y)

X_test = np.linspace(0, 5, 200).reshape(-1, 1)
y_pred = model.predict(X_test)

plt.scatter(X, y, alpha=0.7, label="data")
plt.plot(X_test, y_pred, lw=2, label="linear fit")
plt.legend(); plt.title("Linear regression on noisy sine");
print("k =", model.coef_[0], "  b =", model.intercept_)
print("RÂ² =", r2_score(y, model.predict(X)))
```

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC

X, y = make_classification(n_samples=400, n_features=2, n_redundant=0,
                           n_clusters_per_class=1, random_state=3)

clf = SVC(kernel="rbf", probability=True).fit(X, y)

print("predict:", clf.predict(X[:5]))
print("decision_function:", clf.decision_function(X[:5]))
print("predict_proba:", clf.predict_proba(X[:5]))
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=600, centers=4, cluster_std=0.60, random_state=0)

kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=30)
plt.scatter(*centers.T, c="red", s=120, marker="X")
plt.title("K-means clustering"); plt.show()
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

digits = load_digits()
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, digits.target, test_size=0.25, random_state=1)

mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=400,
                    random_state=1).fit(X_train, y_train)

print("Accuracy:", accuracy_score(y_test, mlp.predict(X_test)))
```

```python
import numpy as np, matplotlib.pyplot as plt, seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.datasets import load_iris

iris_raw = load_iris(as_frame=True)
iris_df = iris_raw.frame
X_iris = iris_df.drop('target', axis=1) if 'target' in iris_df.columns else iris_df.iloc[:, :-1]
```

```python
inertias = []
K = range(1, 11)
for k in K:
    inertias.append(KMeans(n_clusters=k, random_state=0).fit(X_iris).inertia_)

plt.plot(K, inertias, "o-"); plt.xticks(K)
plt.xlabel("k"); plt.ylabel("Inertia"); plt.title("Elbow curve"); plt.show()
```

```python
silh = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=0).fit(X_iris)
    silh.append(silhouette_score(X_iris, km.labels_))

plt.plot(range(2, 11), silh, "o-")
plt.xlabel("k"); plt.ylabel("Silhouette score"); plt.title("Silhouette analysis"); plt.show()
```

```python
eps_values = np.linspace(0.2, 1.0, 9)
scores, clusters = [], []

for eps in eps_values:
    db = DBSCAN(eps=eps, min_samples=5).fit(X_iris)
    labels = db.labels_
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    clusters.append(n_clusters)
    scores.append(np.nan if n_clusters <= 1 else silhouette_score(X_iris, labels))

fig, ax1 = plt.subplots()
ax1.plot(eps_values, scores, "o-", label="silhouette")
ax1.set_xlabel("eps"); ax1.set_ylabel("silhouette score")
ax2 = ax1.twinx()
ax2.plot(eps_values, clusters, "s--", color="orange", label="clusters")
ax2.set_ylabel("clusters found")
fig.legend(); plt.title("DBSCAN parameter sweep"); plt.show()
```
